```python
crops=0
points=0
while ('true'):
  print('welcome to the game')
  print('plant a seed')
  print('water the plant')
  print('harvest the plant')
  print('quit')
  choice=input('enter your choice')
  if choice =='1':
    crops+=1
    print('you planted a seed☘☘')
  elif choice =='2':
    if crops>=1:

      print('you watered the plant🥛')
    else:
      print('you need to plant a seed first')
  elif choice =='3':
    if crops >=1:
      points=points+50

      print('you harvested the plant🌾🌾')
    else:
      print('you need to plant a seed first')
  elif choice =='4':
    print('you quit the game✋')
    print("total points",points)
    break
  else:
    print('invalid choice')
```
