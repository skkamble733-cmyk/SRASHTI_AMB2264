```python
cgpa={1:8.9,2:5.6,4:6.7,7:9.1,8:5.3}
keylist=list(cgpa.keys())
vallist=list(cgpa.values())
minval=min(vallist)
minindex=vallist.index(minval)
minkey=keylist[minindex]
print(minkey)
```

```python
cgpa={1:8.9,2:5.6,4:6.7,7:9.1,8:5.3}
keylist=list(cgpa.keys())
vallist=list(cgpa.values())
minval=min(cgpa.values())
print("value=",minval,"key-->",keylist[vallist.index(minval)])
```

```python
cgpa={1:8.9,2:5.6,4:6.7,7:9.1,8:5.3}
print("key having min value", min(cgpa,key=cgpa.get))
print("key having max value", max(cgpa,key=cgpa.get))
```

```python

cars={'brand':'ford','model':'mustang','year':1964}
print(cars)
cars['color']='red'
print("after adding (color)",cars)
print(cars.keys())
print(cars.values())
print(cars.items())
print("Is key 'model' present", 'model' in cars.keys())
```
