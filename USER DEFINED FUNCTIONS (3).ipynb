```python
def fact(n):
  fact=1
  if (n>0):
   for i in range(1,n+1):
    fact=fact*i
   print("factorial of ",n,'is',fact)
  else:
   print('enter a valid number')
n=int(input("enter a number"))
fact(n)
```
