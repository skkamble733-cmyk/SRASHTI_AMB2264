```python
import random
inventory={}
total_money=0
fish_prices={'rohu':12,'catla':18,'sinchu':25}
print('1 catch fish')
print('2 view inventory')
print('3 sell fish')
print('4 exit')
while True:
  choice=int(input('choose an option'))
  if choice==1:
    fish=random.choice(list(fish_prices.keys()))
    price=fish_prices[fish]
    inventory[fish]=inventory.get(fish,0)+1
    print(f"you caught a {fish}!")
    total_money+=price
  elif choice==2:
    if inventory=={}:
     print("inventory:")
    for fish,count in inventory.items():
      print(f"{fish}:{count}")
      print(f"total money:{total_money}")
  elif choice==3:
    if inventory=={}:
      print("no fish to sell")
    else:
      earnings=0
      for fish, count in inventory.items():
        earnings+=fish_prices[fish]*count
      total_money+=earnings
      inventory.clear()
      print('you sold all fish for $ {earnings}')
  elif choice==4:
    print(f'final money earned:${total_money}')
    print('thanks for playing')
    break
  else:
    print('invalid choice')
```
