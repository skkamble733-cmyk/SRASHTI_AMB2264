```python
s='today is 20-jan-2026 monday'
bc = 0 # Initialize as integer for counting blank spaces
vc = 0 # Initialize as integer for counting vowels
cc = 0 # Initialize as integer for counting consonants

for i in s:
    if i == ' ':
        bc += 1
    elif i.lower() in 'aeiou': # Make vowel check case-insensitive
        vc += 1
    else:
        # Only count if it's an alphabet character, otherwise it could be a digit or symbol
        if i.isalpha():
            cc += 1

print(f"Number of blank spaces: {bc}")
print(f"Number of vowels: {vc}")
print(f"Number of consonants: {cc}")
```

```python
s='today is 20-jan-2026 monday'
b=v=c=d=o=0
for i in s:
  if i==' ':
    b=b+1
  elif i=='a' or i=='e' or i=='i' or i=='o' or i=='u':
    v=v+1
  elif i.isalpha():
      c=c+1
  elif i.isdigit():
      d=d+1
  else:
    o=o+1
print(b,v,c,d,o)
```
