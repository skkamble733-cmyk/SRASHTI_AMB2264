```python
soil_moisture=[45,60,30,55,40]
print("soil moisture levels in different fields:")
for moisture in soil_moisture:
  print(moisture,"%")
```

```python
print("\nfields that need irrigation")
for moisture in soil_moisture:
  if moisture <50:
    print(moisture,"%-irrigation needed")
  else:
    print(moisture,"%-sufficient moisture")
```

```python
print("\ntotal fields checked:",len(soil_moisture))
```

```python
soil_moisture.append(65)
print("after adding new field moisture:")
print(soil_moisture)
print("\n total fields checked:",len(soil_moisture))
```

```python
soil_moisture.sort()
print("moisture levels after sorting:")
print(soil_moisture)
```

```python
print("minimum moisture:",min(soil_moisture),"%")
print("maximum moisture:",max(soil_moisture),"%")
```

```python
soil_moisture.remove(30)
print("after removing 30% moisture value:")
print(soil_moisture)
print("total fields checked:",len(soil_moisture))
```

```python
crops=["rice","wheat","millet","maize","pulses"]
water_need=[300,200,100,150,80]
print("length",len(crops))
print("range",range(len(crops)))
print("crop and water requirement:")
for i in range(len(crops)):
  print(crops[i],"-",water_need[i],"mm")
```

```python
print("crop and water requirement:")
for crop,water in zip(crops,water_need):
  print(crop,"-",water,"mm")
```

```python
print("crops needing high water(>180mm):")
for i in range(len(water_need)):
   if water_need[i]>180:
      print(crops[i])

```

```python
for crop,water in zip(crops,water_need):
  if water>180:
    print(crop)
```

```python
crops=["rice","wheat","maize","pulses"]
yield_tons=[3.5,4.0,2.8,1.5]
average_yield=sum(yield_tons)/len(yield_tons)
print("average crop yield:",average_yield,)
```

```python
max_water=max(water_need)
index=water_need.index(max_water)
print("crop with highest water requirement:",crops[index])
```

**zip method**

```python
max_yield=max(yield_tons)
index=yield_tons.index(max_yield)
print("crop having maximum yield is:",crops[index])
```

```python
print("high yield crops:")
for i in range(len(crops)):
  if yield_tons[i]>average_yield:
    print(crops[i])
```

```python
print("high yield crops:")
# The variable 'yield_tons' was overwritten to a float (1.5) in the kernel state.
# Re-initializing it to its original list form to allow iteration.
yield_tons_data = [3.5, 4.0, 2.8, 1.5] # Using a temporary name to avoid confusion.

# The variable 'crops' in the kernel state is correct (['rice', 'wheat', 'maize', 'pulses']).
# The 'crop' in the zip argument refers to the last string value, not the list 'crops'.
for crop_name, current_yield in zip(crops, yield_tons_data):
 if current_yield > average_yield:
  print(crop_name)
```

```python
crops=["rice","wheat","maize","pulses"]
yield_tons=[3.5,4.0,2.8,1.5]
average_yield=sum(yield_tons)/len(yield_tons)
print("average crop yield:",average_yield)
print("crops with high yield:")
for crop,yields in zip(crops,yield_tons):
  if yields>average_yield:
    print(crop)
```
