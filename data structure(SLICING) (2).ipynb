```python
numbers=[10,20,30,40,50,60]
for i in numbers:
    print(i+2)
print(numbers)
```

```python
n=[]
for i in numbers:
  n.append(i+2)
print("new_list",n)
print("origial list",numbers)
```

```python
n=[i+2 for i in numbers]
print("new_list",n)
print("original list",numbers)
```

```python
n=[]
for i in numbers:
  if i%4==0:
    n.append(i)
print("numbers divided by 4",n)
print("original list",numbers)
```

```python
n = [i for i in numbers if i%4==0]
print("numbers divided by 4", n)
print("original list", numbers)
```

```python
numbers=[10,21,30,41,50,61]
n=["even" if i%2==0 else "odd" for i  in numbers]
print(n)
```
