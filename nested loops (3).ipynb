**if else**

```python
crops=["rice","wheat","maize"]
print("number of crops growm:",len(crops))
if"maize" in crops:
  print("maize is present in list of crops")
if"onion"in crops:
  print("onion is present in list of crops")
else:
  print("onion is not present in list of crops")
crops.remove("wheat")
print(crops)
crops.count("wheat")
index_maize=crops.index("maize")
print("maize is at index:",index_maize)
print(crops.index("maize"))
```

```python
index_rice=crops.index("rice")
print("rice is at index:",index_rice)
print(crops.index("rice"))
crops.insert(2,"sugarcane")
print("after inserting sugarecane")
print(crops)
sugarcane_count=crops.count("sugarcane")
print("sugarcane is grown",sugarcane_count,"times")
crops.pop()
print("after removing last crop")
print(crops)
crops.pop()
print("after removing first crop")
print(crops)
```
